```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
  'employee_id': [101, 102, 103, 104, 105, 106, 107, 102, 108],
  'name': [' Ana Garcia ', 'BOB SMITH', 'Clara_Dupont', 'diego ruiz',
            'Emma Johnson', 'LUCA ROSSI', 'sara chen', 'BOB SMITH', 'Yuki Tanaka'],
  'department': ['Sales', 'sales', 'Tech', 'TECH', 'HR','Hr', 'Tech', 'sales', 'HR'],
  'salary': ['50000', '62000', '75000', 'N/A', '58000',
            '71000', 'N/A', '62000', '66000'],
  'start_date': ['2021-03-01', '2019-07-15', '2020-11-30', '15/04/2022',
                 '2023-01-10', '2018-06-01', '2022-09-20', '2019-07-15', '2020-05-14'],
  'performance': [85, 92, np.nan, 78, 95, np.nan, 88, 92, 71],
  'email': ['ana@company.com', 'bob@company.com', 'clara@company.com',
            'diego@company.com', 'emma@company.com', 'luca@company.com',
            'sara@company.com', 'bob@company.com', 'yuki@company.com']
})
df
```

```python
#Challenge 1 — easy
#Run a full audit on the raw dataset before touching anything. 
#Print the number of duplicates, the number of nulls per column, and the data type of each column. 

print(df.info())
print(df.isnull().sum())
print(df.duplicated().sum())

print(df[df.duplicated()])

df = df.drop_duplicates().reset_index(drop=True)
```

```python
df
```

```python
#Challenge 2 — medium
#Clean the entire dataset in the correct order: duplicates first, then text, then data types, then dates, then nulls. 
#Print the clean df at the end. Order matters — think about why before you start.

df['name']=df['name'].str.strip()
df['name']=df['name'].str.replace('_',' ')
df['name']=df['name'].str.title()

df['department'] = df['department'].str.strip().str.title()

print(df[['name','department']])


```

```python
df
```

```python
df['department']=df['department'].str.replace('Hr','HR')
```

```python
df
```

```python
#Challenge 3 — hard
#After cleaning: find the average salary and average performance score per department. 
#Which department pays the most? Which performs best? Present both results sorted highest to lowest.

df['salary']=df['salary'].replace('N/A',np.nan)

df['salary']=df['salary'].astype(float)

df['salary']=df.groupby('department')['salary'].transform(
    lambda x: x.fillna(x.mean())
)

df['performance'] = df['performance'].fillna(df['performance'].median())

df

```

```python
df['start_date'] = pd.to_datetime(df['start_date'], format='mixed', dayfirst=True)
df['start_year'] = df['start_date'].dt.year
df['years_at_company'] = 2024 - df['start_date'].dt.year
print(df[['name', 'start_date', 'years_at_company']])
```

```python
df
```

```python
#Challenge 4 — boss level
#Add a new column called seniority based on years_at_company: less than 2 years = 'Junior', 2 to 4 years = 'Mid', more than 4 years = 'Senior'. 
#Then count how many employees are in each seniority band per department



conditions = [
    df['years_at_company'] < 2,           # Junior
    (df['years_at_company'] >= 2) & (df['years_at_company'] <= 4),  # Mid
    df['years_at_company'] > 4            # Senior
]


choices = ['Junior', 'Mid', 'Senior']


df['seniority'] = np.select(conditions, choices, default='Unknown')


result = df.groupby(['department', 'seniority'])['employee_id'].count()
print(result)




```
